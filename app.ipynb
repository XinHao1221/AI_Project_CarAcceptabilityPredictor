```python

from flask import Flask, render_template, request
from werkzeug.wrappers import Request, Response
import pickle
import numpy as np
```

```python
model = pickle.load(open('car_acceptability.pkl', 'rb'))
```

```python
app = Flask(__name__)

@app.route('/')
def main():
    return render_template("homepage.html")
```

```python
@app.route('/predict', methods=['POST'])
def home():
    category = request.form['slct1']
    price = request.form['price']
    maintenanceCosts = request.form['maintenance_costs']
    noOfDoors = request.form['no_of_doors']
    noOfPassengers = request.form['no_of_passengers']
    trunkSize = request.form['trunk_size']
    safety = request.form['safety']

    value1 = 0
    value2 = 0
    value3 = 0
    value4 = 0
    value5 = 0
    value6 = 0
    
    carCategoryValue = 0
    
    if(category == 'Hatchback'):
        carCategoryValue = 1
    elif(category == 'Sedan'):
        carCategoryValue = 2
    elif(category == 'SUV'):
        carCategoryValue = 3
    
    #heck user option for car price
    if(category == 'Hatchback'):
        if(price == 'low_price'):
            value1 = 1
        elif(price == 'med_price'):
            value1 = 2
        elif(price == 'high_price'):
            value1 = 3
        elif(price == 'vhigh_price'):
            value1 = 4

    elif(category == 'Sedan'):
        if(price == 'low_price'):
            value1 = 1
        elif(price == 'med_price'):
            value1 = 2
        elif(price == 'high_price'):
            value1 = 3
        elif(price == 'vhigh_price'):
            value1 = 4

    elif(category == 'SUV'):
        if(price == 'low_price'):
            value1 = 1
        elif(price == 'med_price'):
            value1 = 2
        elif(price == 'high_price'):
            value1 = 3
        elif(price == 'vhigh_price'):
            value1 = 4

    # Check user option for car maintenance costs
    if(category == 'Hatchback'):
        if(maintenanceCosts == 'low_maintenance_costs'):
            value2 = 1
        elif(maintenanceCosts == 'med_maintenance_costs'):
            value2 = 2
        elif(maintenanceCosts == 'high_maintenance_costs'):
            value2 = 3
        elif(maintenanceCosts == 'vhigh_maintenance_costs'):
            value2 = 4

    elif(category == 'Sedan'):
        if(maintenanceCosts == 'low_maintenance_costs'):
            value2 = 1
        elif(maintenanceCosts == 'med_maintenance_costs'):
            value2 = 2
        elif(maintenanceCosts == 'high_maintenance_costs'):
            value2 = 3
        elif(maintenanceCosts == 'vhigh_maintenance_costs'):
            value2 = 4

    elif(category == 'SUV'):
        if(maintenanceCosts == 'low_maintenance_costs'):
            value2 = 1
        elif(maintenanceCosts == 'med_maintenance_costs'):
            value2 = 2
        elif(maintenanceCosts == 'high_maintenance_costs'):
            value2 = 3
        elif(maintenanceCosts == 'vhigh_maintenance_costs'):
            value2 = 4

    # Check user option for car trunk size
    if(category == 'Hatchback'):
        if(trunkSize == 'small_trunk_size'):
            value3 = 1
        elif(trunkSize == 'med_trunk_size'):
            value3 = 2
        elif(trunkSize == 'large_trunk_size'):
            value3 = 3

    elif(category == 'Sedan'):
        if(trunkSize == 'small_trunk_size'):
            value3 = 1
        elif(trunkSize == 'med_trunk_size'):
            value3 = 2
        elif(trunkSize == 'large_trunk_size'):
            value3 = 3

    elif(category == 'SUV'):
        if(trunkSize == 'small_trunk_size'):
            value3 = 1
        elif(trunkSize == 'med_trunk_size'):
            value3 = 2
        elif(trunkSize == 'large_trunk_size'):
            value3 = 3
            
    # Check user option for no of Doors
    if(category == 'Hatchback'):
        if(noOfDoors == '2doors'):
            value4 = 1
        elif(noOfDoors == '3doors'):
            value4 = 2
        elif(noOfDoors == '4doors'):
            value4 = 3
        elif(noOfDoors == '5more'):
            value4 = 4

    elif(category == 'Sedan'):
        if(noOfDoors == '2doors'):
            value4 = 1
        elif(noOfDoors == '3doors'):
            value4 = 2
        elif(noOfDoors == '4doors'):
            value4 = 3
        elif(noOfDoors == '5more'):
            value4 = 4

    elif(category == 'SUV'):
        if(noOfDoors == '2doors'):
            value4 = 1
        elif(noOfDoors == '3doors'):
            value4 = 2
        elif(noOfDoors == '4doors'):
            value4 = 3
        elif(noOfDoors == '5more'):
            value4 = 4

    # Check user option for no of passengers
    if(category == 'Hatchback'):
        if(noOfPassengers == '2passengers'):
            value5 = 1
        elif(noOfPassengers == '4passengers'):
            value5 = 2
        elif(noOfPassengers == '5OrMorePassengers'):
            value5 = 3

    elif(category == 'Sedan'):
        if(noOfPassengers == '2passengers'):
            value5 = 1
        elif(noOfPassengers == '4passengers'):
            value5 = 2
        elif(noOfPassengers == '5OrMorePassengers'):
            value5 = 3

    elif(category == 'SUV'):
        if(noOfPassengers == '2passengers'):
            value5 = 1
        elif(noOfPassengers == '4passengers'):
            value5 = 2
        elif(noOfPassengers == '5OrMorePassengers'):
            value5 = 3

    # Check user option for car safety feature
    if(category == 'Hatchback'):
        if(safety == 'low_safety'):
            value6 = 1
        elif(safety == 'med_safety'):
            value6 = 2
        elif(safety == 'high_safety'):
            value6 = 3

    elif(category == 'Sedan'):
        if(safety == 'low_safety'):
            value6 = 1
        elif(safety == 'med_safety'):
            value6 = 2
        elif(safety == 'high_safety'):
            value6 = 3

    elif(category == 'SUV'):
        if(safety == 'low_safety'):
            value6 = 1
        elif(safety == 'med_safety'):
            value6 = 2
        elif(safety == 'high_safety'):
            value6 = 3
            
    if(value1 == 0 or value2 == 0 or value5 == 0):
        return render_template('homepage.html')
       

    arr = np.array([[value1, value2, value3, value4, value5, value6]])
    result = [0, carCategoryValue, value1, value2, value3, value4, value5, value6]
    result[0] = model.predict(arr)
    return render_template('result.html', data=result)
```
